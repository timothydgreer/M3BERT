This script visualizes the outputs (original frames, input frames, and reconstructed frames) from the pretraining prediction process.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
```

```python
example_loc = "../ccmi-bert/outputs/example1"

input_frames = np.load(example_loc+"_input.npy").T
original_frames = np.load(example_loc+"_original.npy").T
reconstructed_frames = np.load(example_loc+"_reconstructed.npy").T
```

```python
viridis = cm.get_cmap('viridis', 256)

def plot_examples(cmap, data, save_loc=None):
    fig, axs = plt.subplots(1, 3, figsize=(15, 3), constrained_layout=True)
    for [ax, i] in zip(axs, data):
        psm = ax.pcolormesh(i, cmap=cmap, vmin=0, vmax=1)
        fig.colorbar(psm, ax=ax)
    if(save_loc):
        plt.savefig(save_loc)
    else:
        plt.show()

plot_examples(viridis, [input_frames, original_frames, reconstructed_frames])
```

### output to images

```python
for i in range(10):
    example_loc = "../ccmi-bert/outputs/example" + str(i)
    input_frames = np.load(example_loc+"_input.npy").T
    original_frames = np.load(example_loc+"_original.npy").T
    reconstructed_frames = np.load(example_loc+"_reconstructed.npy").T
    plot_examples(viridis, [input_frames, original_frames, reconstructed_frames], save_loc="../images/pretrain_poc/example"+str(i)+".png")
```
